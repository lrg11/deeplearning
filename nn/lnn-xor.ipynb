```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[1, 3, 3],[1,4,3],[1,1,1],[1,0,2]])
y = np.array([[1],[1],[-1],[-1]])
# 权值初始化， 3 行 1 列  ，取值范围 -1 到 1
w = (np.random.random([3,1]) - 0.5) *  2
print(w)
# 学习率
lr = 0.12

out = 0

def updateNN():
    global x,y,w,lr
    out = np.dot(x,w) # shape : (4, 1)
    w_c = lr *  (x.T.dot(y-out)) / int(x.shape[0]) # 取到行数，也就是 数据个数
    w = w + w_c
```

```python
for _ in range(100):
    updateNN()
    #print(i)
    #out = np.sign(np.dot(x,w)) # 计算当前输出
    #if(out == y).all(): # 实际输出全部等于期望，模型收敛
    #    print("finished")
    #    print('epoch:', i)
    #    break
x1 = [3, 4]
y1 = [3, 3]
# 负样本
x2 = [1, 0]
y2 = [1, 2]

# 计算分界线的斜率和截距
k = -w[1] / w[2]
d = -w[0] / w[2]
print('k=', k, 'd=', d)
xdata = (0, 5)

plt.figure()
plt.plot(xdata, xdata * k + d, 'r')
plt.scatter(x1,y1, c='b')
plt.scatter(x2, y2, c ='y')
plt.show()
```
